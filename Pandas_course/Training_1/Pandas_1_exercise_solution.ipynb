```python
# Exercise 1: import required libraries
import pandas as pd
import numpy as np
```

```python
# Exercise 2: select column Barcode, MMS ID and Title
df = pd.read_csv('data/items.csv', dtype=str)
df_bar_mms_tit = df[['Barcode', 'MMS Id', 'Title']].copy()
df_bar_mms_tit
```

```python
# Exercise 3: select only the first 100 rows
df_bar_mms_tit.iloc[:100]
```

```python
# Exercise 4: create a new column to indicate if MMS ID is duplicated (option: first is kept)
df_bar_mms_tit['duplicated'] = df_bar_mms_tit['MMS Id'].duplicated()
df_bar_mms_tit
```

```python
# Exercise 5: Suppress duplicates
dedup_df = df_bar_mms_tit.drop_duplicates(subset='MMS Id').copy()

# To check it=> must be 0:
print(dedup_df['duplicated'].sum())
```

```python
dedup_df
```

```python
# Exercise 6: Add quotes to the title '"' The title => "The title"
def update_title(title):
    return '"' + title + '"'

title = 'Le prix de la dépendance comparaison des...'
update_title(title)
```

```python
with_quotes_df = dedup_df.copy()

# In a new column
with_quotes_df['Title_updated'] = with_quotes_df['Title'].apply(update_title)
with_quotes_df
```

```python
with_quotes_df = dedup_df.copy()

# In 'Title' column
with_quotes_df['Title'] = with_quotes_df['Title'].apply(update_title)
with_quotes_df
```

```python
with_quotes_df = dedup_df.copy()

# In a new column with lambda function
with_quotes_df['Title_updated'] = with_quotes_df['Title'].apply(lambda title: f'"{title}"')
with_quotes_df
```
