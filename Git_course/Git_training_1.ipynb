## 2. Goals of git
1. Versioning of the code
2. Sharing code
3. Collaborative tool
4. Workflow management

## 3. Some basic commands

Copy a repository to local:

```
git clone git@github.com:Swiss-Library-Service-Platform/python_training.git
```

Update a local repository:
```
git pull
```

Get current state of git repository
```
git status
```

Reset a repository to the last commit
```
git reset --hard
```


## 4. Update data on Github
A Github account is required to update code on github.

A SSH key is also required. You can generat it with PuTTYgen application on Windows.



Explain to the system what files are useful to add to the repository
```
git add <file_path>
```

Validate the change
```
git commit -m "Explanation about the change"
```

Transmit the change to the remote repository
```
git push
```
