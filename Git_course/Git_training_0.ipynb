# Git training
How to use git for training

## 1. Clone repository
Open a terminal:


Use following command line => will copy the repository
```
git clone https://github.com/Swiss-Library-Service-Platform/python_training.git
```

## 2. Pull updates
Go into the repository folder
```
cd python_training
```

Fetch the updates
```
git pull
```

## 3. Reset local changes
If you make changes into the repository, it could not be possible to pull the updates from the server. In this case it can be useful to reset the local repository.

```
git reset --hard
```
But be careful => you will lost local changes. For example if don't make copies of the exercise notebooks, these one will be restored to the original state -> empty.
