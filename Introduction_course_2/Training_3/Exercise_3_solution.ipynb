```python
# 1.Get the number of barcodes in "barcodes1.txt" and "barcodes2.txt"
with open('data/barcodes1.txt', 'r') as f:
    barcodes1 = f.readlines()
barcodes1 = barcodes1[1:] # remove header
barcodes1 = [barcode[:-1] for barcode in barcodes1] # remove break line

with open('data/barcodes2.txt', 'r') as f:
    barcodes2 = f.readlines()
barcodes2 = barcodes2[1:] # remove header
barcodes2 = [barcode[:-1] for barcode in barcodes2] # remove break line

print(f'barcodes1.txt: {len(barcodes1)}\nbarcodes2.txt: {len(barcodes2)}')
```

```python
# 2. In which file is "DSVN1133713" barcode, "barcode1.txt" or "barcode2.txt" ?
print(f'Is "DSVN1133713" in "barcode1.txt": {"DSVN1133713" in barcodes1}')
print(f'Is "DSVN1133713" in "barcode2.txt": {"DSVN1133713" in barcodes2}')
```

```python
# 3. From file "barcode1.txt" delete all barcodes starting by "DSVN" and save the new list in a "barcode3.txt"
barcodes3 = [barcode for barcode in barcodes1 if barcode.startswith('DSVN') is False] # filter
barcodes3_with_header = ['Barcode'] + barcodes3 # add header
barcodes3_txt = '\n'.join(barcodes3_with_header) # add break lines
with open('data/barcodes3.txt', 'w') as f:
    f.write(barcodes3_txt)
```

```python
# 4. Add the prefix "OLD-" to all barcodes of the list of barcodes3.txt and save in "barcode4.txt"
with open('data/barcodes3.txt', 'r') as f:
    barcodes3 = f.readlines()
barcodes3 = barcodes3[1:] # remove header
barcodes3 = [barcode[:-1] for barcode in barcodes3] # remove break line

barcodes4 = ['OLD-' + barcode for barcode in barcodes3] # add prefix
barcodes4_with_header = ['Barcode'] + barcodes4
barcodes4_txt = '\n'.join(barcodes4_with_header) # add break lines
with open('data/barcodes4.txt', 'w') as f:
    f.write(barcodes4_txt)
```

```python
# 5. Delete "barcode3.txt"
os.remove('data/barcodes3.txt')
```
