## Comprehension lists
[introduction 1 course - comprehension lists lesson](https://jupyter.slsp.ch/user/shared/tree/shared/Python_training/Introduction_course_1/Lists_comprehension)

```python
# String transformation
# A common problem is break lines at the end of readlines function output.
barcodes_without_breaklines = [barcode.replace('\n', '') for barcode in barcodes]
```

```python
barcodes_without_breaklines
```

```python
# Filter a list
# The condition is checked. If the condition is false, the item will not be in the returned list
[barcode for barcode in barcodes if barcode.startswith('A100')]
```

```python
# Both are possible at the same time
[barcode.replace('\n', '') for barcode in barcodes if barcode.startswith('A100')]
```

## Dictionary

```python
# Get element
dict_sample = {'key1': 'value1',
               'key2': 2,
               'key3': 'value3'}
```

```python
dict_sample['key1']
```

```python
# Check if a key exists
if 'key1' in dict_sample:
    print('"key1" is a key of dict_sample')
else:
    print('"key1" is not a key of dict_sample')
```

```python
# Check if a key exists
if 'key4' in dict_sample:
    print('"key4" is a key of dict_sample')
else:
    print('"key4" is not a key of dict_sample')
```

```python
# We can check keys with "in"
'key3' in dict_sample
```

```python
# We cannot check values with "in"
'value3' in dict_sample
```

```python
# Iterate in Dict means iterate on keys values:
for k in dict_sample:
    print(k)
```

```python
# It is possible to fetch the values too, for example:
for k in dict_sample:
    print(k, ':', dict_sample[k])
```

```python
# It is possible to fetch the values too, for example:
for item in dict_sample.items():
    print(item)
```
